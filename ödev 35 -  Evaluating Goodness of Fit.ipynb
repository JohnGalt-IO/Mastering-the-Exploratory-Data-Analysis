# 2. Weather model

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas  as pd
```

```python
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

```python
weather_df["Temp_Difference"] = weather_df["Temperature (C)"]-weather_df["Apparent Temperature (C)"]
```

```python
X = weather_df[['Humidity', 'Wind Speed (km/h)']]
y = weather_df["Temperature (C)"]
```

```python
X_cons = sm.add_constant(X)
results = sm.OLS(y, X_cons).fit()
results.summary()
```

```python
X = weather_df[['Humidity', 'Wind Speed (km/h)']]
y = weather_df["Temp_Difference"]
X_cons = sm.add_constant(X)
results = sm.OLS(y, X_cons).fit()
results.summary()
```

**When we use temperature as a target variable, the R square value is  around 0.418. On the other hand, when we use temperature difference as a target variable, the R squared value is around 0.288. So for both model, R-Squared values are low. but in the temperature difference target variable is too low.**

```python
weather_df["interaction_humid_wind"] = weather_df['Humidity']*weather_df['Wind Speed (km/h)']
X = weather_df[['Humidity', 'Wind Speed (km/h)',"interaction_humid_wind"]]
y = weather_df["Temperature (C)"]
X_cons = sm.add_constant(X)
results = sm.OLS(y, X_cons).fit()
results.summary()
```

**When we add the interaction, the R-Squared values is increasing a little bit from 0.418 to 0.419**

```python
X = weather_df[['Humidity', 'Wind Speed (km/h)',"Visibility (km)"]]
y = weather_df["Temperature (C)"]
X_cons = sm.add_constant(X)
results = sm.OLS(y, X_cons).fit()
results.summary()
```

```python
weather_df["interaction_humid_wind"] = weather_df['Humidity']*weather_df['Wind Speed (km/h)']
X = weather_df[['Humidity', 'Wind Speed (km/h)',"interaction_humid_wind","Visibility (km)"]]
y = weather_df["Temperature (C)"]
X_cons = sm.add_constant(X)
results = sm.OLS(y, X_cons).fit()
results.summary()
```

- **The both of them R-squared and adj. R-squared values increase from 0.418 to 0.449.**
- **And when we add the interaction variable nothing change in term of R-Squared values**

# 2. House prices model

```python
house_prices_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/house_prices.csv", sep = ";")
house_prices_df.head()
```

```python
miss_val_ind = (house_prices_df.isnull().sum().sort_values(ascending=False)/len(house_prices_df)*100).head(10).index
house_prices_df.drop(miss_val_ind[0:5],axis=1,inplace=True)
house_prices_df["LotFrontage"].fillna(house_prices_df["LotFrontage"].mean(),inplace=True)
house_prices_df.dropna(inplace=True)
house_prices_df.info()
```

```python
obj_ind = house_prices_df.dtypes[house_prices_df.dtypes=="O"].index
for i in obj_ind:
    if house_prices_df[i].nunique()>5:
        house_prices_df.drop(i,axis=1,inplace=True)
```

```python
obj_ind = house_prices_df.dtypes[house_prices_df.dtypes=="O"]
corr_ind = house_prices_df.corr()["SalePrice"].sort_values(ascending=False).index
house_prices_df.drop(corr_ind[11:],axis=1,inplace=True)
complet_df = pd.get_dummies(house_prices_df,drop_first=True)
```

```python
X = complet_df.drop("SalePrice",axis=1)
y = complet_df["SalePrice"]
X_cons = sm.add_constant(X)
results = sm.OLS(y, X_cons).fit()
results.summary()
```

Run your house prices model again and assess the goodness of fit of your model using F-test, R-squared, adjusted R-squared, AIC and BIC.

- **The R squared value is around 0.833. we can say that the model is good to use**
- **F-test is around 78.25**
