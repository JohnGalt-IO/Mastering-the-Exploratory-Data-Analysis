```python
import pandas as pd
house_prices_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/house_prices.csv", sep = ";")
house_prices_df.head()
```

```python
perc_col = (house_prices_df.isnull().sum())/len(house_prices_df)*100
```

```python
 perc_col.sort_values(ascending=False).head(20)
```

```python
perc_ind = perc_col.sort_values(ascending=False).head(20).index[]
```

```python
df_drop_ind = perc_col.sort_values(ascending=False).head(20).index[:5]
```

```python
house_prices_df.drop(df_drop_ind,axis=1,inplace=True)
```

```python
house_prices_df
```

```python
perc_col.sort_values(ascending=False).head(20)
```

```python
house_prices_df.LotFrontage.fillna(house_prices_df.LotFrontage.mean(),inplace=True)
```
