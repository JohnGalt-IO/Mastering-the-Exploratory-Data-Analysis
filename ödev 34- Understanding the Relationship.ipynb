# 2. Weather model

In this exercise, you'll work with the historical temperature data from the previous lesson. To complete this assignment, submit a link in the gist file to the Jupyter notebook containing your solutions to the following tasks:

```python
import pandas as pd
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

```python
weather_df["Temp_Difference"] = weather_df["Temperature (C)"]-weather_df["Apparent Temperature (C)"]
```

```python
X = weather_df[['Humidity', 'Wind Speed (km/h)']]
y = weather_df["Temperature (C)"]
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
```

```python
X_cons = sm.add_constant(X_train)
results = sm.OLS(y_train, X_cons).fit()
results.summary()
```

- **p values of features lower than 0.05 it means that the features are important**
- **The sign and value of features does not change.** 
- **There is negative relation between exploratory variable with target variable**

```python
weather_df["interaction_humid_wind"] = weather_df['Humidity']*weather_df['Wind Speed (km/h)']
```

```python
X = weather_df[['Humidity', 'Wind Speed (km/h)',"interaction_humid_wind"]]
y = weather_df["Temperature (C)"]
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
```

```python
X_cons = sm.add_constant(X_train)
results = sm.OLS(y_train, X_cons).fit()
results.summary()
```

# 3. House prices model

In this exercise, you'll interpret your house prices model. To complete this assignment, submit a link in the gist file to the Jupyter notebook containing your solutions to the following tasks:

```python
house_prices_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/house_prices.csv", sep = ";")
house_prices_df.head()
```

```python
miss_val_ind = (house_prices_df.isnull().sum().sort_values(ascending=False)/len(house_prices_df)*100).head(10).index
house_prices_df.drop(miss_val_ind[0:5],axis=1,inplace=True)
house_prices_df["LotFrontage"].fillna(house_prices_df["LotFrontage"].mean(),inplace=True)
house_prices_df.dropna(inplace=True)
house_prices_df.info()
```

```python
obj_ind = house_prices_df.dtypes[house_prices_df.dtypes=="O"].index
for i in obj_ind:
    if house_prices_df[i].nunique()>5:
        house_prices_df.drop(i,axis=1,inplace=True)
```

```python
obj_ind = house_prices_df.dtypes[house_prices_df.dtypes=="O"]
corr_ind = house_prices_df.corr()["SalePrice"].sort_values(ascending=False).index
house_prices_df.drop(corr_ind[11:],axis=1,inplace=True)
complet_df = pd.get_dummies(house_prices_df,drop_first=True)
```

```python
complet_df.head()
```

```python
X = complet_df.drop("SalePrice",axis=1)
y = complet_df["SalePrice"]
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
```

```python
X_cons = sm.add_constant(X_train)
results = sm.OLS(y_train, X_cons).fit()
results.summary()
```
